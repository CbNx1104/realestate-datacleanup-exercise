#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
prices_fuenlabrada = sb.loc[sb['level5'] == "Fuenlabrada",'price']
norm_prices_fuenlabrada = (prices_fuenlabrada - prices_fuenlabrada.min()) / prices_fuenlabrada.std()
prices_leganes = sb.loc[sb['level5'] == "Leganés",'price']
norm_prices_leganes = (prices_leganes - prices_leganes.min()) / prices_leganes.std()
prices_getafe = sb.loc[sb['level5'] == "Getafe",'price']
norm_prices_getafe = (prices_getafe - prices_getafe.min()) / prices_getafe.std()
prices_alcorcon = sb.loc[sb['level5'] == "Alcorcón",'price']
norm_prices_alcorcon = (prices_alcorcon - prices_alcorcon.min()) / prices_alcorcon.std()

fig, axes = plt.subplots(nrows=2, ncols=2)
ax0, ax1, ax2, ax3 = axes.flatten()
n_bins = 6
ax0.hist(norm_prices_fuenlabrada, bins = n_bins, alpha = 1)
ax0.set_title('Fuenlabrada')

ax1.hist(norm_prices_leganes, bins = n_bins, alpha = 1)
ax1.set_title('Leganés')

ax2.hist(norm_prices_getafe, bins = n_bins, alpha = 1)
ax2.set_title('Getafe')

ax3.hist(norm_prices_alcorcon, bins = n_bins, alpha = 1)
ax3.set_title('Alcorcón')

plt.show()
```

**NOTE:** In Fuenlabrada, Leganés and Alcorcón the mayority of the prices are about one standard deviation from the minimum price, the more the prices go farther from the minimun the frequency gets lower. But for Getafe the mayority is distrubited almost equaly up to 3 standard deviations from the minimum price, which means there are a few properties with really high price in the sample located in Getafe causing a big disparity in the prices, this could be causing a deviation.

```python
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon

# Map centered on (60 degrees latitude and -2.2 degrees longitude)
# Latitude, longitude
map = Map(basemap=basemaps.OpenStreetMap.Mapnik,center = (40, 3.7), zoom = 5, min_zoom = 5, max_zoom = 50)
map
```
